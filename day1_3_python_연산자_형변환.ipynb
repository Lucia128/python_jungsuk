## 1.3 할당 연산자 (❗연산자 + 문법 혼합)
- =  +=  -=  *=  /=  //=  %=  **=  &=  |=  ^=  <<=  >>=  
- = → ❌ 메서드 없음. 연산자가 아니라 문장(statement)
- 대입은 “객체의 행동”이 아니라 “이름의 재배치”다.
- += → __ iadd__ → 못찾으면 __ add__
- 나머지도 동일 패턴 (__ isub__, __ imul__, … )
-  x += y 이 항상 x = x + y 와 같은건 아니다.(mutable 주의)

```python
x = y = 1 # 자바와 동일
print(f"{x=}")
print(f"{y=}")
# print(x=y=1) # 에러
```

```python
a = 10    # 대입이 아니라 바인딩. 10이란 객체를 a에 바인딩

# 자바와 달리 좌변이 여러개 일 수 있다. 이걸 하나의 메서드로 통일 불가능.
x, y = 1, 2
lst[0] = 10
obj.attr = 10
y, x = x, y
```

```python
# 속성 대입 또는 인덱스 대입의 경우 메서드 호출
class A : pass
obj = A()
lst = [1,2,3]
obj.x = 10   # obj.__setattr__("x", 10) 속성 대입
lst[0] = 10  # lst.__setitem__(0, 10)   인덱스 대입
```

## 1.4 논리 연산자 (❗연산자 아님)  
- and,  or,  not
- ❌ 매직 메서드 없음
- ❌ 오버로딩 불가
- ⭕ short-circuit 평가

- 🧵 왜 and / or 는 메서드가 아닌가?  
  단락 평가(short-circuit) 때문이고(함수나 메서드는 모든 인자를 확인)  
  연산자 자체가 “제어 흐름”에 가깝기 때문


```python
A = True
B = False
print(A or B)   # A가 True면 A, 아니면 B

A = False
B = True
print (A and B) # A가 False면 A, 아니면 B
```

## 1.7 동일성 연산자 (문법)
- is, is not
- 객체 주소(identity) 비교
- 오버로딩 불가

```python
str1 = str(123)
str2 = str(123)
print(str1 is str2)      # False. 동일성(identity) 비교
print(str1 is not str2)  # True
```

```python
str1 = str(123)
str2 = str(123)
print(str1==str2) # 동등성(equality)비교
print(str1!=str2)
```

## 1.14 Walrus 연산자 - :=  
- :=는 “값을 계산하면서 동시에 이름에 바인딩”하는 연산자다.
- 할당 + 식   if (n := len(data)) > 10:
- 오버로딩 불가
- 새 변수 생성 가능
- 식(expression) 안에서 사용 - 반드시 괄호 쳐야함. 
- 문(statement) 아님.   <---  대입문이 아니게 하려고 =대신 :=사용
- python 3.8+

```python
# y := 10   # 오류
# (y := 10) + 5   # 15 괄호 반드시 필요
print((y := 10) + 5)
```

```python
# line = input()
# while line != "":
#     print(line)
#     line = input()

# 위의 코드를 아래와 같이 개선
while (line := input()) != "":
    print(line)
```

# 2. 형변환


## 2.1 명시적 타입 변환 (Explicit casting) - 개발자가 직접 함수 호출로 변환

| 변환        | 방법           | 예시                       |
| --------- | ------------ | ------------------------ |
| → int     | `int(x)`     | `int("10") → 10`         |
| → float   | `float(x)`   | `float("3.14") → 3.14`   |
| → complex | `complex(x)` | `complex(1, 2) → (1+2j)` |


```python
print(int(3.9))   # 3 (버림)
int("3.9") # ValueError
```

## 2.2 문자열 변환
| 변환    | 방법                 | 예시               |
| ----- | ------------------ | ---------------- |
| → str | `str(x)`           | `str(10) → "10"` |
| 포맷 변환 | `format`, f-string | `f"{x:.2f}"`     |


## 2.3 컬렉션 타입 변환
| 변환 대상   | 방법                | 결과        |
| ------- | ----------------- | --------- |
| → list  | `list(iterable)`  | 리스트       |
| → tuple | `tuple(iterable)` | 튜플        |
| → set   | `set(iterable)`   | 중복 제거     |
| → dict  | `dict(pairs)`     | key-value |


```python
print(list("abc"))        # ['a', 'b', 'c'] 문자열을 문자 리스트로 변환
print(set([1, 1, 2]))     # {1, 2}     리스트에서 중복이 제거된 set으로 변환
print(dict([("a",1)]))    # {'a': 1}   튜플이 담긴 리스트를 dict로 변환
```
