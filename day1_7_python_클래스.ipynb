## 1.2 객체의 생성과 초기화 - __ new__, __ init__
- 하나의 클래스에 한개 씩만 존재해야. 오버로딩x -> *args, **kwargs사용 또는 @classmethod로
- __ new__ : 객체가 생성될 때 호출
- __ init__ : 객체가 초기화될 때 호출.
- 클래스이름() -> __ call__   
-> (1) __ new__ 호출 : 객체 생성  
-> (2) __ init__호출 : 객체 초기화  

```python
class A :
    def __new__(cls):
        print(f"{cls=}의 객체가 생성됨 by __new__()")
        return super().__new__(cls)
        
    def __init__(self):
        print(f"{self=}가 초기화됨 by __init__()")

a = A()  # 객체를 생성(__new__)하고 초기화(__init__)
print(a) # <__main__.A object at 0x107f2af90>
        
```

### 생성자 오버로딩 대신 클래스 메서드로 처리

## 1.3 __ repr__ : 객체를 문자열로 표현하는 방법을 정의하는 메서드

```python
class Point :
    # 객체를 초기화하는 메서드
    def __init__(self, x=0, y=0):
        print("객체 초기화. by __init__()")
        self.x = x # 속성 x추가, 인스턴스 변수
        self.y = y 
    
    # 객체를 문자열로 표현할 때 호출되는 메서드
    def __repr__(self):
        return f"Point({self.x}, {self.y})" # 객체를 문자열로 표현해서 반환
     
p = Point() # 매개변수 기본값 사용
print(p) # (0, 0)
print(p.__repr__()) # 위와 동일.
p.x = 1 # 객체의 속성 x를 변경
p.y = 2
print(p)
```

## 1.4 __ eq__ : 객체의 동등성을 확인하는 메서드. True, False를 결과로 반환

```python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __eq__(self, other):
        if not isinstance(other, Point):
            return False
        return self.x == other.x and self.y == other.y
    
print(Point(3,5)==Point(1,2))
print(Point(3,5)==Point(3,5))
print(Point(1,2)=="(1, 2)")
```

## 1.6 인스턴스 변수와 클래스 변수
- 인스턴스 변수는 객체마다 갖는 객체 고유의 변수
- 클래스 변수는 같은 클래스로 생성된 객체가 공유하는 변수
- 클래스 변수와 인스턴스 변수의 이름이 같을 때는 인스턴스 변수가 우선

```python
class Card:
    width = 100    # 클래스 변수
    height = 200
    
    def __init__(self, kind, num):
        self.kind = kind # 인스턴스 변수
        self.num  = num
        
    def __repr__(self):
        return f"({self.kind}, {self.num})"
    
c = Card("H", 1)
c2 = Card("D", 5)
print(c)

Card.height = 50  # 클래스 변수의 변경
c.width = 200 # 인스턴스 변수의 추가. 클래스 변수의 변경x
print(f"{c.width=}")   # c의 인스턴스 변수 width(새로 추가됨)
print(f"{c2.width=}")  # c2의 클래스 변수 width
print(f"{c.height=}")  
print(f"{c2.height=}")
```

```python
class B:
    count = 0

    def inc(self):
        B.count += 1 # 클래스 변수의 값을 1증가
        # type(self).count += 1 # 위와 동일
        
b = B()
b.inc()
b.inc()
b.inc()

B.count
```

```python
class Config:
    timeout = 30   # 기본값 (클래스 변수)

    def __init__(self, timeout=None):
        if timeout is not None:
            self.timeout = timeout

c1 = Config()   # 값이 지정되지 않으면 기본값(클래스 변수)이 사용됨
c2 = Config(10) # 값이 지정되면 지정된 값이 인스턴스 변수로 추가됨.

print(f"{c1.timeout=}")  # 30 (클래스 변수). 인스턴스 변수가 추가되지 않아 클래스 변수가 사용됨
print(f"{c2.timeout=}")   # 10 (인스턴스 변수)
```

## 1.7 인스턴스 메서드
- 인스턴스에 연결된 함수 -> 메서드
- 첫번째 메개변수로 인스턴스(self)를 받음. self는 호출시 자동 주입됨

a.foo
- 1. A.__ dict__['foo'] 찾음
- 2. 이 함수는 descriptor (__ get__ 있음)
- 3. function.__ get__(a, A) 호출
- 4. bound method 객체 생성 - 클래스 내에 정의된 함수가 호출시 객체와 연결된 메서드가 된다.
     인스턴스 메서드 = 인스턴스(__ self__) + 함수  
     클래스 메서드   = 클래스(__ self__) + 함수

```python
class A:
    def foo(self, x): # 인스턴스 메서드. self는 반드시 첫번째 매개변수이여야
        print(f"{self=}, {x=}")

a = A()
print(A.foo) # A.foo는 클래스에 정의된 함수(function)
print(a.foo) # a.foo는 객체 a에 연결된 메서드(bound method)
print(a.foo.__func__) # 메서드 안에 속성으로 함수가 저장되어 있음
print(a.foo.__self__) # 함수와 연결된 객체
```

```python
class A:
    def foo(self, x): # 인스턴스 메서드. self는 반드시 첫번째 매개변수이여야
        print(f"{self=}, {x=}")

a = A()
a.foo(10) # foo(self, x)호출. self는 자동대입. 여기서는 a가 self. 
A.foo(a, 10) # 위의 문장과 동일. self=a, x= 10
```

## 1.9 Singleton패턴
- 하나의 객체만 생성해서 재사용하는 디자인 패턴

```python
class Singleton:
    _instance = None
    def __new__(cls):
        if(cls._instance is None):
            cls._instance = super().__new__(cls)
        return cls._instance
        
    @classmethod    
    def getInstance(cls):
        if cls._instance is None:
            cls._instance = cls()
        return Singleton._instance
    

s = Singleton.getInstance()
s2 = Singleton.getInstance()
print(s is s2)
print(s==s2)
```

### 속성 탐색 우선순위(중요)  
단순화하면, descriptor먼저 -> 인스턴스 뒤지기 -> 클래스 뒤지기
1. data descriptor (class) 변수 먼저  
2. instance.__ dict__       객체 먼저  
3. non-data descriptor (class)  
4. class.__ dict__  클래스는 나중에  
5. 부모 클래스 (__ mro__)  


```python
class A:
    @property   # 디스크립터. a.x처럼 읽기만 가능(read only). 대입블가.
    def x(self): # a.x처럼 점연산자(.)사용시 호출되는 메서드
        return 10

a = A()
result = a.x # 10  속성 x를 읽을 때 x()가 자동 호출됨. 
print(f"{result=}")
a.x = 100 # 쓰기는 에러. AttributeError: property 'x' of 'A' object has no setter
```

### descriptor로 쓰기도 가능하게 만들기: setter 추가


```python
class A:
    @property    # 읽기 가능
    def x(self):
        return self._x
    @x.setter    # 쓰기 가능
    def x(self, value):
        self._x = value
        
a = A()
# a.x # 아직은 x가 없음.
a.x = 10  # 객체 a에 속성 x가 추가됨. 속성 x의 setter가 있어서 쓰기도 가능
a.x # 10
```

```python
# 양수 또는 0인지 검증(validation)하는 descriptor
class Positive:
    def __get__(self, obj, owner):
        return obj._x

    def __set__(self, obj, value):
        if value < 0:  # 음수면 ValueError발생
            raise ValueError
        obj._x = value

class A:
    x = Positive() # descriptor생성
    
a = A()
# x가 descriptor이면, descriptor의 __set__() 호출
# descriptor가 아니면, a객체의 속성(__dict__) 중에 찾음
a.x = 10  # 양수를 대입했으므로 __set__()의 검증 통과
a.x = -10 # 음수라서 ValueError 
```

```python
class A:
    @property
    def x(self):
        return self._x
    
    @x.setter    # 쓰기 가능
    def x(self, value):
        self._x = value
        
    @x.deleter  # 이게 있으면 속성의 삭제도 가능
    def x(self):
        del self._x
        
a = A()
a.x = 10 # 속성 x 추가. 이 문장이 있어야 삭제 가능
del a.x # OK
a.x     # AttributeError: 'A' object has no attribute '_x'
```

```python
# [주의] 무한 재귀에 빠지지 않게 내부변수 사용(_x)
@x.setter
def x(self, value):
    # self.x = value   # 무한 재귀
    self._x = value  # 반드시 내부 변수 사용
```

## 1.12 속성의 읽기, 변경하기, 삭제하기
- Python의 철학: “모든 동작은 가로챌 수 있어야 한다”
- 객체의 속성에 접근할 때 점(.)연산자는 다음과 같이 처리

obj.attr  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;↓  
__ getattribute__  점(.)연산자에 의해 제일 먼저 호출됨. 이 메서드가 descriptor를 호출할지를 판단  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;↓  
data descriptor?  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;↓  
instance dict / slots?  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;↓  
class dict (descriptor?)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;↓  
MRO  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;↓  
AttributeError → __ getattr__  

```python
class A:
    def __getattribute__(self, name):
        return 42

    def __getattr__(self, name):
        return 82

class B:
    def __getattr__(self, name):
        return 82

a = A()
a.x      # 42  __getattribute__(a, x) -> 42
a.y      # 42  __getattribute__(a, y) -> 42

b = B()
b.x      # 82  __getattribute__(a, x)가 없어서 -> __getattr__(a,x) -> 82
```

```python
class A:
    x = 10 # 클래스 속성(클래스 변수)
    
    def __getattribute__(self, name):
        print("__getattribute__", name)
        return super().__getattribute__(name) # 여기서 순서대로 다 뒤지는 로직처리

    def __getattr__(self, name):      # __getattr__은 에러 발생시 처리하기 위한 것
        print("__getattr__", name)
        return 99

a = A()
a.x  # __getattribute__호출. 인스턴스에 속성 x가 없어서 클래스 속성에서 찾음.
a.y  # __getattribute__호출. 인스턴스에 속성 x가 없어서 클래스 속성에서도 못찾아서 __getattr__호출
```

## 1.13 slots - 고정된 개수의 속성만 저장. 동적 추가x

- 메모리를 절약하고 동적 속성 생성을 제한
- __ slots__ → instance dict 자체 제거


### slot을 이용하면 왜 메모리가 줄어드는가?
- dict 제거
- 해시 테이블 사용x
- 포인터 수 감소
 - 대량 객체에서 효과 큼

```python
class A:
    def __init__(self):
        self.x = 1
        self.y = 2
        
a = A()
a.z = 3   # __dict__에 동적으로 속성 추가 가능
a.__dict__
```

```python
class B:
    __slots__ = ('x', 'y')

b = B()
b.x = 1
b.y = 2
# b.z = 3 # 동적 속성추가 불가. AttributeError: 'B' object has no attribute 'z'

# 객체 b에 __dict__가 존재하지 않음
b.__dict__  # AttributeError: 'B' object has no attribute '__dict__'
```

```python
class Base:
    __slots__ = ('a',)

class Child(Base):
    __slots__ = ('b',)

class GrandChild(Child):  # 손자는 slot사용 안하면? __dict__사용
    pass

c = Child()
c.a = 1
c.b = 2
# c.__dict__ # AttributeError: 'Child' object has no attribute '__dict__'

g = GrandChild()
g.a = 1
g.b = 2
g.__dict__ # OK
```

# 2. 상속
- 기존 클래스의 속성(데이터)과 동작(메서드)을 재사용·확장하는 메커니즘
- 다른 클래스(부모)의 속성을 모두 물려받는 새로운 클래스를 작성
- 다중 상속 지원. 여러 클래스를 부모로 가지면, 동시에 여러 클래스의 속성을 물려받을 수 있음.


```python
# object : 모든 클래스의 최고 조상
class Animal(object): # object는 생략가능
    def speak(self):
        print("sound")

class Dog(Animal):
    pass

a = Animal()
a.speak()

d = Dog()
d.speak() # 조상으로부터 상속받은 메서드를 호출

isinstance(d, Animal) # True
isinstance(d, Dog)    # True
issubclass(Animal, object) # True. Animal은 object의 자손
issubclass(Dog, Animal)    # True. Dog는 Animal의 자손  
issubclass(type(d), Animal) # True. 윗 줄과 동일
```

## 2.2 다중 상속 - 여러 조상으로부터 상속받기

### 속성 탐색 순서
- 1. obj.__dict__
- 2. 클래스 MRO 순서대로 (Method Resolution Order)
- 3. 못 찾으면 AttributeError

```python
class A: pass
class B: pass
class C(A, B): pass     # 다중 상속 A와 B를 상속
class D(C): pass

# 속성을 찾는 순서. D -> C -> A -> B -> object
D.mro() # [D, C, A, B, object]
```

```python
class A:
    def f(self): print("A")

class B:
    def f(self): print("B")

class C(A, B): # 다중 상속 - A와 B를 상속
    pass

c = C()
print(C.mro()) # C -> A -> B -> object의 순서로 속성을 찾음.
print(c.f())   # A의 f()를 호출. B보다 A의 순서가 먼저
```

## 2.2 오버라이딩 - 조상의 메서드를 자손에서 재정의
- super()로 조상의 메서드 호출 가능 - super()가 반환하는 조상은 MRO에 의해 결정
- 특정 조상 메서드를 직접 호출도 가능(추천x)

```python
class Animal:
    def speak(self):
        print("sound")

class Dog(Animal):
    def speak(self):    # 오버라이딩. 조상의 메서드를 덮어쓴다.
        super().speak() # 조상의 speak()를 호출. super()는 MRO에 의해 결정되는 조상
        print("bark")

a = Animal()
a.speak() # sound

d = Dog()
d.speak() # bark. 자손에 정의된 메서드가 우선
```

## 2.3 추상 클래스(abstract class)
- 미완성 클래스, 미완성 설계도 -> 추상 메서드(미완성 메서드)를 하나 이상 보유한 클래스
- 공통 인터페이스를 정의하고
- 파생 클래스에서 구현을 강제할 수 있어
- 코드 설계가 단단해지고 일관성을 유지할 수 있습니다.

### 추상 클래스의 조건
- abc.ABC를 상속하고
- @abstractmethod 데코레이터를 붙인 메서드를 넣으면  
→ 자식 클래스에서 반드시 해당 메서드를 구현해야  
- 미구현 추상 메서드가 있으면 객체 생성 시 TypeError가 납니다.

```python
from abc import ABC, abstractmethod # Abastract Base Calss

class Animal(ABC):   # 추상 클래스
    @abstractmethod  # 추상 메서드에 붙여야
    def sound(self): # 선언부만 있고 구현부는 없는 추상 메서드
        pass

class Dog(Animal):  # 추상클래스를 상속 - 추상 클래스의 모든 추상 메서드를 구현해야 객체 생성가능
    def sound(self): # 추상 메서드의 구현 - 몸통 만들어주기
        print("bark!!!")

# a = Animal()  # TypeError: 추상 클래스는 인스턴스화 불가
dog = Dog()   # OK
dog.sound()
```

## 2.4 Protocol (PEP 544)
- 객체가 어떤 메서드나 속성을 가지고 있는지”만을 기준으로 타입을 판단.
- 명시적으로 상속할 필요없이, 필요한 메서드/속성만 존재하면 해당 타입으로 간주됨(duck typing)

### Protocol이 필요한 이유
- 파이썬은 원래 동적 타입 언어라서 런타임에만 타입이 결정됩니다.
- 문제점
1. 타입 안정성을 확보하기 어렵다  
2. 코드 분석/자동완성/에러 검출이 힘들다

```python
from typing import Protocol

class Greeter(Protocol):  # Greeter를 Protocol로 정의
    def greet(self) -> str: ...

class Korean:    # Protocol을 상속받지 않음.
    def greet(self) -> str: # Protocol에 정의된 것과 같은 메서드를 보유
        return "안녕하세요!"

class English:
    def greet(self) -> str:
        return "Hello!"

class Spanish:    # Greeter프로토콜을 구현x. 즉, greet()가 없음.
    def hello(self) -> str:
        return "ola!"


# Greeter프로토콜을 구현한 클래스의 객체를 매개변수로 받음.
def say_hi(g: Greeter) -> None:
    print(g.greet())

# say_hi(Spanish()) # AttributeError: 'Spanish' object has no attribute 'greet'
say_hi(Korean())
say_hi(English())
```

## 2.5 다형성(polymorphism)
- 같은 이름, 다른 형동. 메서드 오버로딩으로 가능
- 덕 타이핑. 타입에 관계없이 동일한 메서드 제공하면 OK. Protocol로 체크

```python
def make_sound(animal):
    print(animal.sound())

class Bird:
    def sound(self):
        return "Tweet"

class Dog:
    def sound(self):
        return "Bark"
    
class Fly:
    pass

make_sound(Bird())
make_sound(Dog())
make_sound(Fly()) # AttributeError: 'Fly' object has no attribute 'sound'
```

```python
from typing import Protocol

class Sounder(Protocol): # Sounder를 Protocol로 정의
    def sound(self) -> str:  # str을 반환하는 메서드(함수 타입 힌트)
        pass

def make_sound(animal:Sounder):
    print(animal.sound())

class Bird:
    def sound(self):
        return "Tweet"

class Dog:
    def sound(self):
        return "Bark"
    
make_sound(Bird())
make_sound(Dog())
```

## 2.6  클래스 데코레이터

- 함수 대신 클래스에도 데코레이터 함수를 처리할 수 있다.
- 이때는 함수대신 클래스를 받도록 구성한다.

```python
# 클래스에 붙일 수 있는 데코레이터 정의
def class_deco(cls):
    cls.deco = True  # 클래스 변수 추가
    return cls

@class_deco # 클래스 데코레이터를 붙여서 클래스를 확장
class A:
    pass

A.deco
```

### 클래스 데코레이터를 이용해서 불변 클래스만들기

```python
def immutable(cls):
    def new_setattr(self, name, value):
        raise AttributeError("not set attribute ")
    cls.__setattr__ = new_setattr # 새로운 setter로 대체
    return cls

@immutable
class A:
    a = 3 # 클래스 변수
    b = 4
    
    
a = A()
a.a = 10  # 변경하면 예외 발생
```
