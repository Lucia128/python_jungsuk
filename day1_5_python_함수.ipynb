### 함수형 언어 - 순수 함수 vs 비순수 함수

## 1.2 return문 - 처리결과를 반환

- return 문을 사용해서 처리결과를 반환한다. 하나의 값만 반환 가능. 여러개 반환하면 튜플로 묶임
- 처리 결과가 없을 때는 return만 적는다. 이때 None이 반환됨(함수는 항상 값을 반환)
- 함수 내의 문장을 수행하다가 return문을 만나면 수행을 중단하고 호출한 곳으로 되돌아간다.

```python
def add(a,b) : # 함수의 정의
    return a+b # 반환값

def func():
    return      # 아무것도 반환안하면 None이 반환됨

# 함수를 호출하고 그 결과를 result에 저장
result = add(3,5) # 함수의 호출
print(result)
print(func())
```

```python
def func(a,b,c) :
    return a, b, c  # 여러 개를 반환. 하나의 튜플로 묶임

result = func(1,2,3)
print(f"{result=}")
print(type(result))
```

## 1.3 함수 호출하기

- 함수를 정의한 문장을 읽으면, 함수객체가 생성되어 실제 함수 이름과 함께 네임스페이스에 등록
- 함수는 호출할 수 있으므로 callable이며, 호출연산자 () 또는 __ call__ 메소드를 사용해서 호출


### 함수 이름도 식별자이므로 이름으로 조회하면 함수 객체를 가져온다.

```python
def add(a,b) : # 함수의 정의
    return a+b # 

print(add) # 함수 객체를 출력
print(callable(add)) # True. 함수는 호출할 수 있는 callable. callable객체는 __ call__을 가지고 있음
print(add.__call__)
print(callable(100)) # False. 정수값은 호출할 수 없으므로
# print((100).__call__) # 에러. 100은 callable이 아니라 __ call__이 없음.

result = add(3,5) # 호출연산자 ()로 함수를 호출. 호출연산자는 피연산자에 대해 __ call__을 호출
print(f"{result=}")
```

[참고] callable객체의 종류  
| 종류             | 예시                  | 호출 방식   | 내부 메서드         |
| -------------- | ------------------- | ------- | -------------- |
| 함수 (function)  | `def f(): ...`      | `f()`   | `__call__`     |
| 람다 함수          | `lambda x: x+1`     | `f()`   | `__call__`     |
| 메서드 (bound)    | `obj.m()`           | `()`    | `__call__`     |
| 메서드 (unbound)  | `Class.m`           | `(obj)` | `__call__`     |
| 클래스            | `MyClass()`         | 생성자 호출  | `__call__`     |
| 인스턴스 (functor) | `obj()`             | `()`    | `obj.__call__` |
| 내장 함수          | `len(x)`            | `()`    | C 구현           |
| 내장 메서드         | `list.append`       | `()`    | C 구현           |
| 제너레이터 함수       | `def f(): yield`    | `f()`   | `__call__`     |
| 코루틴 함수         | `async def f()`     | `f()`   | `__call__`     |
| partial        | `functools.partial` | `()`    | `__call__`     |
| 데코레이터 결과       | `@decorator`        | `()`    | `__call__`     |

## 1.6 기본값 인자와 키워드 인자

## 1.7 가변 인자, 가변 키워드 인자
- 가변 인자와 가변 키워드 인자는 ‘몇 개가 올지 모르는 인자’를 하나의 객체로 묶어 받는 문법 설탕

### 가변 위치 인자 *args

- 위치 인자를 개수 제한 없이 받음
- 함수 안에서는 튜플(tuple)

```python
def f(*args):  # 가변인자. 입력된 인자를 튜플로 묶어서 받는다.
    print(type(args)) # <class 'tuple'>
    print(args) # 튜플(모든 인자)을 출력

f()        # ()
f(1,2)     # (1, 2)
f(1, 2, 3) # (1, 2, 3)
```

### 가변 키워드 인자 **kwargs
- 키워드 인자를 개수 제한 없이 받음
- 함수 안에서는 딕셔너리(dict)

```python
def f(**kwargs):
    print(type(kwargs)) # <class 'dict'>
    print(kwargs)

f()
f(a=1)
f(a=1, b=2) # {'a': 1, 'b': 2}
# f(1,2) # TypeError: f() takes 0 positional arguments but 2 were given
```

### 인자의 순서 규칙
- 1. 기본값 인자
- 2. *args
- 3. 키워드 전용 인자
- 4. **kwargs


```python
def f(*args, **kwargs): # 가변인자(tuple)와 키워드 인자(dict)를 같이 사용. 순서에 주의
    print(f"{args} {kwargs}")

f(1,2)
f(x=3, y=4)
f(1, 2, x=3, y=4)
# f(x=3, y=4, 1, 2) # 에러. 순서에 주의
```

```python
# * 이후의 인자는 반드시 키워드 인자로만 전달해야
def f(a, b, *, c, d):  
    print(f"{a=}, {b=}, {c=}, {d=}")

f(1, 2, c=3, d=4)  # OK
# f(1, 2, 3, 4)      # 에러

arg1 = (10,20)
arg2 = {'c': 30, 'd': 40}

# *, **를 이용해서 구조분해
f(*arg1,**arg2) # *, **를 꼭 붙여야함. arg1, arg2가 4개의 인자로 분해해서 대입됨
```

## 1.8 중첩 함수
- 함수 안에 정의된 함수. 중첩의 횟수는 거의 제한 없음.
- 일반적으로 함수 내부에서만 사용
- return문으로 함수 외부로 반환해서 외부에서도 사용 가능

```python
def outer():
    x = 10 # 외부함수 outer()의 지역변수
    def inner(): # 일반적으로 함수 내부에서만 사용. return문으로 외부에 반환 가능
        print("inner()")
        return x   # 외부 함수의 지역변수 사용 가능
    return inner # 내부 함수를 반환

outer()() # outer()는 내부에 중첩된 함수 inner를 반환. 
```

```python
def outer():
    x = 10 # 외부함수 outer()의 지역변수
    def inner(): # 일반적으로 함수 내부에서만 사용. return문으로 외부에 반환 가능
        x = 20 # 내부 함수의 지역변수
        print("inner()의 지역변수: ", locals())

        return x   # 내부 함수의 지역변수. 이름이 같을 때는 가까운 것을 사용
    print("outer()의 지역변수: ", locals())
    return inner # 내부 함수를 반환

# outer()는 내부에 중첩된 함수 inner를 반환. 
outer()() # outer()() -> inner() -> 20
```

### nonlocal과 global
- nonlocal : 외부 함수의 변수와 내부 함수에서 변경할 때 붙인다. 안붙여도 참조가능
- global   : 함수 내부에서 전역 변수를 사용할 때 붙인다. 
- L → E → G → B의 순서로 조회  
Local  
Enclosing   ← nonlocal  
Global     ← global  
Builtins  

```python
def outer():
    x = 10 
    def inner():
        x = 20 # 외부 함수의 지역변수와 같은 이름의 내부 함수의 지역변수
        print(f"{x=}") # 이름이 같을 때 가까운 영역(scope)의 변수를 의미
        return x
    return inner

outer()() # outer()는 내부에 중첩된 함수 inner를 반환. 
```

```python
def outer():
    x = 10 
    def inner():
        nonlocal x # x가 외부 함수의 x임을 의미
        print(f"{x=}") # 외부 함수의 x
        return x
    return inner
print(outer.__closure__)
outer()() # outer()는 내부에 중첩된 함수 inner를 반환. 
```

```python
x = 10 # global변수
def outer():
    x = 20 # 외부 함수의 지역변수
    def inner():
        global x # x가 전역 변수임을 의미
        print(f"{x=}") # 전역변수 x
        return x
    return inner

outer()() # outer()는 내부에 중첩된 함수 inner를 반환. 
```

## 1.9 클로져(closure)
- 외부 스코프의 변수를 기억하는 중첩 함수(외부변수 사용안하면 클로져가 아님)
- 변수는 값이 아니라 참조(cell) 로 캡처
- 수정하려면 nonlocal 필요
- 반복문과 함께 쓰면 늪에 빠지기 쉬움
- 상태 은닉에 매우 강력한 도구
- 클로저 변수는 cell object에 저장. 프레임 종료 후에도 힙에 유지

```python
def outer(x):
    def inner(y):
        return x + y   # x는 외부 변수(자유 변수)
    return inner

f = outer(10)
print(f(3))  # f는 outer가 종료된 후에도 outer의 x를 사용가능

print(f.__closure__)
print(f.__code__.co_freevars) # inner의 자유변수 출력.

```

## 1.10 함수 데코레이터
- 함수를 받아서, 기능을 추가한 새로운 함수로 바꿔주는 함수(데코레이터 패턴)

```python
def hello() : 
    print("hello")
    
def my_decorator(func):
    def wrapper():
        print("before")
        func()
        print("after")
    return wrapper

hello()
hello2 = my_decorator(hello) # hello -> hello2
hello2()
```

```python
def my_decorator(func):
    def wrapper():
        print("before")
        func()
        print("after")
    return wrapper

@my_decorator    # hello()에 my_decorator를 적용
def hello():
    print("hello")

hello() # 자동으로 데코레이터가 동작해서 새로운 함수가 반환 & 호출
```

```python
def my_decorator1(func):
    def wrapper():
        print("before1")
        func()
        print("after1")
    return wrapper

def my_decorator2(func):
    def wrapper():
        print("before2")
        func()
        print("after2")
    return wrapper

@my_decorator2    
@my_decorator1    # 가까운쪽부터 먼저 적용된다.
def hello():
    print("hello")

hello() # 자동으로 데코레이터가 동작해서 새로운 함수가 반환 & 호출
```

### 인자가 있는 경우의 데코레이터

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("before")
        result = func(*args, **kwargs)
        print("after")
        return result
    return wrapper

@my_decorator
def add(a, b):
    """a, b를 더해서 반환 """ # docstring
    return a + b

print(add(3,5))
print(add.__name__) # add가 아니라 wrapper
print(add.docstring) # wrapper
```

### functools.wraps는 필수
- 데코레이터를 적용해도 함수 이름(__name__), 함수설명(docstring)을 유지

```python
from functools import wraps

def my_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper

@my_decorator
def add(a, b):
    """a, b를 더해서 반환 """
    return a + b

print(add.__name__)      # 'add'
print(add.__doc__)       # 원래 docstring
```

## 1.10 재귀 함수
- 함수 내에서 함수 자신을 호출하는 함수
- 반복문과 같은 효과. 반복문보다 성능 낮음. 반복문으로 대체 가능
- 논리적으로 간결한 코드 장성 가능한 경우 사용

```python
def factorial_recursive(n): # 재귀호출 함수. 느리지만 간결
    return n * factorial_recursive(n-1) if n > 1 else 1 # 재귀호출

def factorial_for(n): # 위의 재귀호출 함수와 동등. 더 빠름.
    ret = 1
    for i in range(1, n+1):
        ret *= i
    return ret

print(factorial_for(5))
print(factorial_recursive(5))
```

## 팩토리 계산결과를 캐쉬해서 처리하기
- 재귀함수를 계산호출하면 함수 실행환경을 계속 스택에 구성
- 한번 처리한 것을 메모리에 저장한 후에 다시 호출할 때 있으면 조회한 값을 처리해서 함수 실행환경을 적게 구성

```python
cache = {}

def factorial_rec_cache(n):
    global cache

    if n in cache:
        return cache[n]
    elif n <= 1:
        return 1
    else:
        cache[n] = n * factorial_rec_cache(n-1)
        return cache[n]

    return n * factorial_rec_cache(n-1) if n > 1 else 1
```

```python
%%time
factorial_rec_cache(5)
print(f"{cache=}")
```

```python
%%time
factorial_rec_cache(20)
print(f"{cache=}")
```

## 1.12 람다
- 함수를 간단한 식으로 표현한 것. 이름이 없는 익명 함수
- 한 줄로 정의되고 하나의 식(expression) 만 가질 수 있음.
- 임시로 간단한 함수가 필요할 때
- 다른 함수(map/filter/sorted 등)에 함수를 바로 넘겨줄 때

```python
# lambda 매개변수들 : 식   # 람다 문법. 식의 처리결과를 반환
add = lambda x, y: x + y # x, y를 받아서 더한 결과를 반환하는 람다
print(add(2, 4))  # 6

(lambda x, y: x + y)(3, 5)  # 람다의 정의와 호출을 동시에
```

```python
# map()은 람다를 이용해서 변환
mylist = [1, 2, 3]
result = list(map(lambda x: x * 2, mylist))
print(result)   # [2, 4, 6]
```

```python
# filter()는 람다를 이용해서 조건에 맞는 것만 반환
mylist = [1, 2, 3, 4, 5]
odd = list(filter(lambda x: x % 2 == 1, mylist)) # 홀수만 반환
print(odd)  # [1, 3, 5]
```

```python
# sorted()로 정렬할 때, 정렬 기준을 람다로 제공
names = ['apple', 'kiwi', 'banana']
sorted_names = sorted(names, key=lambda x: len(x)) # 문자열의 길이를 정렬기준으로
print(sorted_names)  # ['kiwi', 'apple', 'banana']
```

```python
# 주의사항 - late binding
funcs = []
for i in range(3):
    funcs.append(lambda: i) # 람다가 실행될때는 이미 i의 값이 2

print([f() for f in funcs])  # [2, 2, 2]

funcs = []
for i in range(3):
    funcs.append(lambda i=i: i) # 외부 변수 i를 캡춰해서 저장
print([f() for f in funcs])  # [0, 1, 2]

```
