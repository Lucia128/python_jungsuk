## 왜 모듈을 쓰는가?

“파일 단위로 로딩되는 객체이며,  
네임스페이스 + 재사용 + 구조화의 핵심 단위”  

👉 파일 분할이 아니라 “의미 분리”
- namespace 분리
- 코드 실행 단위
- 상태 공유
- import 캐시

## 1.1 모듈의 생성과정
- 모듈을 import하면, 내부적으로 생성후 현재 모듈의 namespace에 바인딩  
    1.	ModuleType("foo") # foo모듈 생성  
	2.	foo._ dict__ 생성      #   네임스페이스(__ dict__) 생성  
	3.	exec(code, foo.__ dict__)  
	4.	sys.modules["foo"] = foo <---- sys.modules에 저장  

## 1.2 sys.modules
- sys모듈의 dict[str, module] 모듈이름(str)과 모듈객체를 쌍으로 저장
- sys.modules는 모듈 캐시 + 순환 import 방지 장치. 
- 여러번 import해도 한번만 import

```python
import sys
import math
sys.modules['math']  # 모듈이 import되면, 모듈 객체가 생성되어 sys.modules에 저장됨.
```

## 1.3 모듈의 주요 속성

| 속성명            | 타입              | 설명                          |
| -------------- | --------------- | --------------------------- |
| `__name__`     | `str`           | 모듈 이름. 직접 실행되면 `"__main__"` |
| `__file__`     | `str`           | 모듈 파일의 경로                   |
| `__package__`  | `str` 또는 `None` | 패키지 이름 (상대 import에 사용)      |
| `__spec__`     | `ModuleSpec`    | 모듈 로딩에 대한 명세 객체             |
| `__loader__`   | Loader          | 모듈을 로드한 로더 객체               |
| `__dict__`     | `dict`          | 모듈의 네임스페이스                  |
| `__doc__`      | `str` 또는 `None` | 모듈의 docstring               |
| `__cached__`   | `str` 또는 `None` | 컴파일된 파일 `.pyc` 경로           |
| `__builtins__` | `dict`          | 내장 함수/예외 네임스페이스             |
| `__path__`     | `li             |                             |

## 1.4 모듈의 작성

```python
%%writefile calc.py  
# calc.py - calc모듈을 작성
def add(a, b):
    return a + b

PI = 3.14
```

## 1.5 모듈의 import


```python
import sys   # sys모듈을 import
import os, sys, time # 여러 import문을 한줄로. 여러 모듈 import(한줄에 하나가 바람직)
import itertools as tool  # itertools 모듈의 이름을 별명 tools로 바꿔서 사용
from math import sqrt     # math모듈에서 sqrt만 import해서 사용. 모듈이름 생략 가능
from math import sqrt as sq # sqrt를 sq라는 별칭으로 사용
from math import sin, cos # 여러개 import
from math import *        # math모듈의 모든 이름을 import. 모듈이름 없이 사용 가능(바람직x)
```

- __ init__.py는 “그 디렉터리를 하나의 모듈로 만드는 진입 코드” - 자바의 module-info.java

## 2.3 __ init__.py의 역할
- 패키지 인식. __ init__.py <--- 폴더안에 이 파일이 있으면 패키지
- 패키지 초기화 코드 작성

```python
%%writefile __init__.py
# mypackage/__init__.py
print("mypackage imported") # 외부에서 이 패키지를 import할 때 실행됨.
```

```python
# 공개 API 정의할 때 사용
# mypackage/__init__.py
from .a import func1     # 모듈 a.py에서 func1을 import
from .b import func2     # 모듈 b.py에서 func2을 import

# 외부에 어떤 것을 공개할지 명확히 지정
__all__ = ["func1", "func2"]   # 외부에서 import * 하면 func1, func2가 import되게
```

## 2.3 모듈과 패키지의 비교
- 모듈: 파이썬 코드가 들어 있는 하나의 .py 파일
- 패키지: 여러 모듈을 구조적으로 묶은 디렉터리

| 구분            | 모듈 (Module)                              | 패키지 (Package)                                      |
| ------------- | ---------------------------------------- | -------------------------------------------------- |
| 개념            | **하나의 파이썬 파일**                           | **여러 모듈을 묶은 디렉터리**                                 |
| 형태            | `.py` 파일 1개                              | 디렉터리(폴더)                                           |
| 목적            | 관련된 함수, 클래스, 변수의 묶음                      | 관련된 모듈들의 묶음                                        |
| 구성 요소         | 함수, 클래스, 변수, 상수                          | 모듈 + 서브패키지                                         |
| `__init__.py` | 없음                                       | 있음 (Python 3.3+에서는 선택)                             |
| 계층 구조         | 단일 레벨                                    | 계층 구조 가능                                           |
| import 방식     | `import math`<br>`from math import sqrt` | `import numpy.linalg`<br>`from numpy import array` |
| 재사용 단위        | 작음                                       | 큼                                                  |
| 예시            | `math`, `sys`, `os`                      | `numpy`, `pandas`, `matplotlib`                    |
| 파일 시스템 관점     | 파일                                       | 폴더                                                 |
| 내부에 또 다른 것 포함 | ❌                                        | ✅ (모듈/패키지 포함 가능)                                   |
